```python
import pandas as pd

# Load the spreadsheet
pokemon_data  = pd.read_csv('/content/drive/My Drive/Datasets/pokemon.csv')

# Display the first few rows of the dataset
pokemon_data.head()
```

```python

# Selecting relevant columns to display
statsDF  = pokemon_data[['type1','hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'weight_kg']]

statsDF
```

```python
pokemon_type_summary = statsDF.groupby('type1').mean()
pokemon_type_summary.sort_values(by='hp', ascending=False)
```

```python

import plotly.graph_objects as go

categories = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']

fig = go.Figure()

for pokemon_type in pokemon_type_summary.index:
    fig.add_trace(go.Scatterpolar(
          r = pokemon_type_summary.loc[pokemon_type, categories].values.tolist(),
          theta = categories,
          fill = 'toself',
          name = pokemon_type
    ))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, pokemon_type_summary[categories].max().max()]
    )),
  showlegend=True
)

fig.show()
```

 # Types with the highest average stats

*  Attack: Dragon
*  HP: Dragon
*  Speed: Flying
*  Special Defense: Fairy
*  Special Attack: Psychic
*  Defense: Steel




```python

# Calculate the correlation
correlation = pokemon_data['weight_kg'].corr(pokemon_data['defense'])
print("Correlation between weight and defense:", correlation)

# Visualize the correlation
import plotly.express as px
fig = px.scatter(pokemon_data, x='weight_kg', y='defense',
                 trendline="ols", title=f"Correlation: {correlation:.2f}")
fig.show()
```

```python
# Calculate correlations between height/weight and base stats
correlations_height = pokemon_data[['height_m', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].corr()['height_m']
correlations_weight = pokemon_data[['weight_kg', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].corr()['weight_kg']

# Print the correlations
print("Correlations with Height:\n", correlations_height)
print("\nCorrelations with Weight:\n", correlations_weight)

# Visualize correlations with height
fig_height = px.scatter_matrix(pokemon_data, dimensions=['height_m', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'],
                               title="Correlations with Height")
fig_height.show()

# Visualize correlations with weight
fig_weight = px.scatter_matrix(pokemon_data, dimensions=['weight_kg', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'],
                               title="Correlations with Weight")
fig_weight.show()
```


Based on the correlation analysis of height and weight with base stats, we can draw the following conclusions:

# **Height:**
  Height has very weak correlations with all base stats. This suggests that height doesn't play a significant role in determining a Pokemon's combat abilities.

# **Weight:**
 * Weight has a moderate positive correlation with defense, indicating that heavier Pokemon tend to have better defensive capabilities.
* Weight also shows weak positive correlations with HP and attack, suggesting a slight tendency for heavier Pokemon to have higher HP and attack stats.
* However, weight has weak negative correlations with special attack, special defense, and speed. This implies that heavier Pokemon might be slightly less proficient in these areas.

# **Overall:**
 * Weight seems to be a more influential factor than height when it comes to base stats.
 * Heavier Pokemon generally exhibit better defensive capabilities, but might be somewhat slower and less adept at special attacks.
 * Height doesn't appear to be a strong predictor of any particular base stat.
